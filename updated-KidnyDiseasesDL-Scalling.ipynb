```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('tuned-kidney_disease.csv')
X = dataset.iloc[:, 1:25].astype('float32' )
y = dataset.iloc[:, 25]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout



```

```python
# Initialising the ANN
classifier = Sequential()
# Adding the input layer and the first hidden layer i.e. input laye has 11 inputs (neuran), 2nd has 6 neurans
classifier.add(Dense(units = 24, activation='relu',input_dim = 24))

# Adding the second hidden layer
classifier.add(Dense(units = 20, activation='relu'))

# Adding the 3rd hidden layer
classifier.add(Dense(units = 15, activation='relu'))
#============
#classifier.add(Dense(2, activation= 'softmax'))
#classifier.compile(loss= 'binary_crossentropy' , optimizer= 'Adamax' , metrics=[ 'accuracy' ])
#========
# Adding the output layer
classifier.add(Dense(units = 1, activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
model_history=classifier.fit(X_train, y_train,validation_split=0.25, batch_size = 1, epochs = 20)
#model_history=classifier.fit(X_train, y_train, batch_size = 1, epochs = 20)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Convert to float
df['Predicted Float'] = df['Decision on >0.5'].apply(lambda x: float(str(x).strip('[]')))

# True labels and predicted probabilities
actual = df['Actual Value']
proba = df['Predicted Float']

# ROC calculation
fpr, tpr, thresholds = roc_curve(actual, proba)
roc_auc = auc(fpr, tpr)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Kidney Disease Prediction')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# Step 1: Convert predicted probabilities from 'Predicted Value' column
df['Predicted Float'] = df['Decision on >0.5'].apply(lambda x: float(str(x).strip('[]')))

# Step 2: Extract true labels and predicted probabilities
actual = df['Actual Value']
proba = df['Predicted Float']

# Step 3: Calculate precision and recall
precision, recall, thresholds = precision_recall_curve(actual, proba)

# Step 4: Plot Precision-Recall Curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='b', lw=2, label=f'Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for Kidney Disease Prediction')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
