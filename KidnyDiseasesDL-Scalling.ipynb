```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('tuned-kidney_disease.csv')
X = dataset.iloc[:, 1:25].astype('float32' )
y = dataset.iloc[:, 25]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout


# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer i.e. input laye has 11 inputs (neuran), 2nd has 6 neurans
classifier.add(Dense(units = 24, activation='relu',input_dim = 24))

# Adding the second hidden layer
classifier.add(Dense(units = 20, activation='relu'))

# Adding the 3rd hidden layer
classifier.add(Dense(units = 15, activation='relu'))
#============
#classifier.add(Dense(2, activation= 'softmax'))
#classifier.compile(loss= 'binary_crossentropy' , optimizer= 'Adamax' , metrics=[ 'accuracy' ])
#========
# Adding the output layer
classifier.add(Dense(units = 1, activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
model_history=classifier.fit(X_train, y_train,validation_split=0.25, batch_size = 1, epochs = 20)
#model_history=classifier.fit(X_train, y_train, batch_size = 1, epochs = 20)
```

```python
predicted=[]
for y in y_pred:
    if y>0.5:
        predicted.append(1)
    else:
        predicted.append(0)
print(predicted)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(predicted,y_test)
score
```
